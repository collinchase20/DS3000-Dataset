######  *FEATURE EXTRACTION: Calculate the Better/Worse/Neutral Standing of a Stock Post 2008

```python
def classify_stock_post_08(df):
    'This function looks at 2008 stocks average closing price post recession and compares it to the average closing'
    'prices before and during the recession to determine if the stock is on a better/worse/neutral trend'
    bwn_post_08 = {}
    
    for ind in df.index:
        
        if df['AVG CLOSING POST 08'][ind] > df["AVG CLOSING DURING 08"][ind] and df['AVG CLOSING POST 08'][ind] > df["AVG CLOSING PRE 08"][ind]:
            bwn_post_08.update({df["Symbol"][ind]: "Better"})
            
        elif df['AVG CLOSING POST 08'][ind] < df["AVG CLOSING DURING 08"][ind] and df['AVG CLOSING POST 08'][ind] < df["AVG CLOSING PRE 08"][ind]:
            bwn_post_08.update({df["Symbol"][ind]: "Worse"})
            
        else:
            bwn_post_08.update({df["Symbol"][ind]: "Neutral"})
            
    return bwn_post_08
```

```python
classifications = classify_stock_post_08(recession_2008)
recession_2008["Standing Post 08"] = recession_2008.Symbol.map(classifications)
recession_2008.head()
```

```python
def get_onewayANOVA(df, V1, V2):
    
    df_1 = 1
    df_2 = len(V2) - 1
    
    
    f_results = stats.f_oneway(df[V1], df[V2])

    f_stat = format(f_results[0],".04f") 
    p_value = format(f_results[1], ".50f")
    
    print("F(%s,%s) = %s, p = %s" % (df_1, df_2, f_stat, p_value)) 
    
    
    descriptives = df[V1].agg(["mean", "sem"])
    print(V1, "mean:", descriptives[0], "\nMSE:", descriptives[1])
    descriptives = df[V2].agg(["mean", "sem"])
    print(V1, "mean:", descriptives[0], "\nMSE:", descriptives[1])
```

**Preparation for Hypothesis #4: Will a stocks performance in the 2008 recession relate to its performance in the COVID-19 recession**

```python
#Set up a dataframe to compare the sectors to the differnce in average closing price before both recessions.
r28s = recession_2008[["Sector", "AVG CLOSING PRE 08"]]
r28s.set_index("Sector")

rcs = recession_covid[["Sector", "AVG CLOSING PRE COVID"]]
rcs.set_index("Sector")

rs = pd.merge(r28s, rcs, left_index = True, right_index= True)

rs["Difference"] = rs["AVG CLOSING PRE 08"] - rs["AVG CLOSING PRE COVID"]
del rs["Sector_y"]
rs
```

```python
from scipy.stats import kruskal
import itertools

def oneway_ANOVA_H4(data, IV, DV):
    #print headers
    print("-----------------------" + "\n" + "ONE-WAY ANOVA RESULTS" + "\n" + "-----------------------"
           + "\n" + "\n" + "\n" + "F-test" + "\n" + "-------");
    
    Basic_Materials = data[DV][data[IV] =="Basic Materials"]
    Communication = data[DV] [data[IV] =="Communication Services"]
    Consumer_Cyclical = data[DV][data[IV] =="Consumer Cyclical"]
    Consumer_Defensive = data[DV][data[IV] =="Consumer Defensive"]
    Energy = data[DV][data[IV] =="Energy"]
    Financial = data[DV][data[IV] =="Financial"]
    Financial_Services = data[DV][data[IV] =="Financial Services"]
    Healthcare = data[DV][data[IV] =="Healthcare"]
    Industrials = data[DV][data[IV] =="Industrials"]
    Real_Estate = data[DV][data[IV] =="Real Estate"]
    Technology = data[DV][data[IV] =="Technology"]
    Utilities = data[DV][data[IV] =="Utilities"]
    
    
    #F-test

    result = stats.f_oneway(Basic_Materials, Communication, Consumer_Cyclical, Consumer_Defensive, 
                            Energy, Financial, Financial_Services, Healthcare, 
                            Industrials, Real_Estate, Technology, Utilities)
    
    fStat = format(result[0], '.4f')
    pValue = format(result[1], '.4f')
    #Degree of Freedom
    df1 = 1
    
    df2 = len(IV) - 1
    
    print("\n" + "\n" + "Assumption Checks" + "\n" 
          + "-------------------" + "\n" + "\n" + "Assumption of Equality of Variances:")
    #Checking for Equality of Variances
    levene_results = stats.levene(Basic_Materials, Communication, Consumer_Cyclical, Consumer_Defensive, 
                            Energy, Financial, Financial_Services, Healthcare, 
                            Industrials, Real_Estate, Technology, Utilities)
    
    if levene_results[1] > .05:
        print("\t" + str(levene_results) + "\n" + "\t" + "Assumption is met. p > .05");
    else:
        print("\t" + str(levene_results) + "\n" + "\t" + "Assumption is violated. p < .05");
    
    #print F-test
    print("\n" + "F(" + str(df1) + "," + str(df2) + ") = " + str(fStat) + ", p = " + str(pValue))
    
    
    #Assumption is Violated so Perform kruskals comparison instead of tukeys
    d = {"basic_materials": Basic_Materials, "communication": Communication, "consumer_cyclical": Consumer_Cyclical, "consumer_defensive": Consumer_Defensive, 
                            "energy": Energy, "financial": Financial, "financial_services": Financial_Services, "healthcare": Healthcare, "industrials":
                            Industrials, "real_estate": Real_Estate, "technology": Technology, "utilities": Utilities }
    
    combo = itertools.combinations(list(d.keys()),2)
    for val in combo:
        stat, p = stats.kruskal(d.get(val[0]), d.get(val[1]))
        print(val[0], "," , val[1], 'Statistics=%.3f, p=%.3f' % (stat, p))
```
